```python
#CÓDIGO OFICIAL PARA CONSOLIDAÇAO DE PROCEDIMENTOS

import pandas as pd

#Criar Data Frame do arquivos de procedimentos
df = pd.read_excel('Procedimentos - julho.xlsx')
dadosProcedimentos = pd.DataFrame(df)

# Usando o método value_counts() para contar a frequência de cada profissional
frequencia_profissionais = dadosProcedimentos['PROFISSIONAL'].value_counts().reset_index()

# Renomeando as colunas para 'PROFISSIONAL' e 'Frequência'
frequencia_profissionais.columns = ['PROFISSIONAL', 'FREQUÊNCIA']


# Juntando os DataFrames com base naa coluna 'PROFISSIONAL'
dadosProcedimentos = dadosProcedimentos.merge(frequencia_profissionais, on='PROFISSIONAL', how='left')

# Removendo duplicatas com base na coluna 'PROFISSIONAL'
dadosProcedimentos = dadosProcedimentos.drop_duplicates(subset=['PROFISSIONAL'])

#Criando o Excel com a consolidação dos procedimentos
output_file = 'Procedimentos Consolidados - 072024.xlsx'
dadosProcedimentos.to_excel(output_file, index=False)

print(dadosProcedimentos)

```

```python
#Criar Data Frame dos arquivos base de cadastros e da Totalidade nominal
df1 = pd.read_excel('Tabela de Tecnicos Enf.xlsx')
df2 = pd.read_excel('Procedimentos Consolidados - fevereiro.xlsx')

#Criar variável Merge_df com resultado do merge pelo Nome
merged_df = pd.merge(df1, df2, on='PROFISSIONAL', how='inner')

#Remover duplicados do Merge
colunas_chave = ['PROFISSIONAL']
merged_df = merged_df.drop_duplicates(subset=colunas_chave)

print(merged_df)
#Criar Arquivo Excel com resultado do Merge
file_path = r'C:\ResultadoMergeTecnicos.xlsx'
merged_df.to_excel(file_path, index = False)
```
