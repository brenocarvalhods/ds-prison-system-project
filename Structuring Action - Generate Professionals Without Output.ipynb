```python
#PRIMEIRA AÇÃO ESTRUTURANTE: CARREGAR DADOS DOS PROFISSIONAIS QUE NAO TIVERAM PRODUÇÃO NO MÊS

#Para isso serão realizadas 04 etapas
#Primeira Etapa: carregar arquivo gerado no Banco com todos os atendimentos realizados no mês e respectivos profisisonais
#E arquivo gerado através da nossa planilha de RH

import pandas as pd

# Carregue o primeiro arquivo Excel (arquivo oriundo do banco de dados com os atendimentos)
dfAtendimentosRealizados = pd.read_excel('Atendimentos - 072024.xlsx')
dfAtendimentosRealizados = pd.DataFrame(dfAtendimentosRealizados)

# Carregue o segundo arquivo Excel (arquivo oriundo da primeira etapa e arquivo de RH, lembrando de alterar para inserir coluna de Profissionais)
dfFileProfissionaisRH = pd.read_excel('Planilha RH - 072024.xlsx')
dfFileProfissionaisRH = pd.DataFrame(dfFileProfissionaisRH)

# Etapa de limpeza dos dados dos dois DataFrames
# Excluir linhas com coluna 'Profissional' em branco
dfAtendimentosRealizados = dfAtendimentosRealizados[dfAtendimentosRealizados['Profissional'].notna()]
dfFileProfissionaisRH = dfFileProfissionaisRH[dfFileProfissionaisRH['Profissional'].notna()]

# Segunda Etapa: fazer um merge entre os dois arquivos e trazer os dados dos profissionais
# que estão na tabela de profissionais das unidades, mas não estão na tabela de atendimentos realizados

# Pesquisa dos nomes da coluna profissional do segundo DataFrame no primeiro DataFrame
nomes_encontrados = dfFileProfissionaisRH['Profissional'].isin(dfAtendimentosRealizados['Profissional'])

# Filtra os nomes que não foram encontrados e inclui as colunas de Unidade e Equipe
nomes_nao_encontrados = dfFileProfissionaisRH.loc[~nomes_encontrados, ['Profissional', 'LOTAÇÃO']]

# Mostra os nomes não encontrados
print("Nomes não encontrados:", nomes_nao_encontrados['Profissional'].tolist())

# Salva o resultado do merge (profissionais não encontrados) em um arquivo Excel com as colunas desejadas
nomes_nao_encontrados.to_excel('Profissionais sem produção - julho.xlsx', index=False)
```
