```python
#Codigo para gerar hiperutilizadores (Privados de Liberdade com mais de 8 atendimentos por categoria profisisonal) na saude.

import pandas as pd

#Criar Data Frame do arquivo de atendimentos
dfAtendimentosRealizados = pd.read_excel('Atendimentos 2023 - Hiper.xlsx')

#Fazendo a limpeza dos dados
dfAtendimentosRealizados = dfAtendimentosRealizados.drop_duplicates(subset=['NOME', 'CNS', 'CPF', 'DT. ATENDIMENTO', 'CBO'])
```

```python
# Filtrar os atendimentos para os CBOs específicos
cbo_especificos = ['MÉDICO']  # Exemplo de códigos CBO para MÉDICO, ENFERMEIRO e PSICÓLOGO CLÍNICO

# Filtra o DataFrame para manter apenas as linhas onde a coluna 'CBO' contém a palavra 'MÉDICO'
df_cbo_filtrado = dfAtendimentosRealizados[dfAtendimentosRealizados['CBO'].str.contains('MÉDICO', case=False, na=False)]

# Contar atendimentos por pessoa
atendimentos_por_pessoa = df_cbo_filtrado['NOME'].value_counts()
pessoas_mais_de_8_atendimentos = atendimentos_por_pessoa[atendimentos_por_pessoa > 8].index

# Filtrar o dataframe para essas pessoas
df_filtrado_final = df_cbo_filtrado[df_cbo_filtrado['NOME'].isin(pessoas_mais_de_8_atendimentos)]

# Adicionar a coluna de contagem de atendimentos
df_filtrado_final['NUM. ATENDIMENTOS'] = df_filtrado_final.groupby('NOME')['NOME'].transform('count')

# Selecionar as colunas desejadas
colunas_desejadas = ['NOME', 'CPF', 'CNS', 'UNIDADE', 'CBO', 'NUM. ATENDIMENTOS']
dfHiperutilizadoresMedicos = df_filtrado_final[colunas_desejadas]

# Remover duplicatas para exibir apenas uma linha por pessoa
# Converta a coluna 'Nome' para maiúsculas
dfHiperutilizadoresMedicos['NOME'] = dfHiperutilizadoresMedicos['NOME'].str.upper()

# Remova as duplicatas com base nas colunas 'Nome' e 'CPF'
dfHiperutilizadoresMedicos = dfHiperutilizadoresMedicos.drop_duplicates(subset=['NOME', 'CPF'])

print(dfHiperutilizadoresMedicos)
```
