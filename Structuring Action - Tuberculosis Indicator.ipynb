```python
import pandas as pd

#Create Data Frame from base files
dfAtendimentosNoMes = pd.read_excel('Atendimentos Gerais - 072024.xlsx')
dfAtendimentosMensaisTB = pd.read_excel('Atendimentos TB - 072024.xlsx')
dfListaCondicaoAtivaTB = pd.read_excel('TB Ativa - 28082024.xlsx')

#Converting everything to uppercase
dfAtendimentosNoMes = dfAtendimentosNoMes.applymap(lambda x: x.upper() if isinstance(x, str) else x)
dfAtendimentosMensaisTB = dfAtendimentosMensaisTB.applymap(lambda x: x.upper() if isinstance(x, str) else x)
dfListaCondicaoAtivaTB = dfListaCondicaoAtivaTB.applymap(lambda x: x.upper() if isinstance(x, str) else x)

#FIRST STEP: RESULT SPREADSHEET WITH WHO WAS SERVED IN THE LAST MONTH AND IS ALSO CONTAINED IN THE ACTIVE CONDITION SPREADSHEET
dfAtendidosAtivos = pd.merge(dfAtendimentosNoMes, dfListaCondicaoAtivaTB, on=['NOME'])
dfAtendidosAtivos = dfAtendidosAtivos.drop_duplicates(subset=['NOME', 'CPF', 'CNS'])

#SECOND STEP: RESULT SPREADSHEET WITH WHO WAS SERVED IN THE LAST MONTH, WITH TB CIAP, AND IS CONTAINED IN THE ACTIVE CONDITION SPREADSHEET
dfAtendidosAtivosTB = pd.merge(dfAtendidosAtivos, dfAtendimentosMensaisTB, on=['NOME'])
dfAtendidosAtivosTB = dfAtendidosAtivosTB.drop_duplicates(subset=['NOME'])

#THIRD STEP: CREATE FILE AND TABS
with pd.ExcelWriter('DGASP - Dados TB - julho.xlsx', engine='openpyxl', mode='w') as writer:
    dfAtendidosAtivos.to_excel(writer, sheet_name='Atendidos e Ativos', index=False)
    dfAtendidosAtivosTB.to_excel(writer, sheet_name='Atendidos e Ativos - CIAP TB', index=False)
```
