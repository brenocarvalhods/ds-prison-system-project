```python
#O codgio abaixo visa carregar alguns DataFrames, realizar o merge entre os mesmos e  gerar algumas
#abas necessarias para a analise de dados de cadastro do sistema prisional.

import pandas as pd

#Carregar arquivos e criar Data Frames dos arquivos base de cadastros e da Totalidade nominal
dfCadastrosBase = pd.read_excel('Cadastros julho.xlsx')
dfCadastrosSeres = pd.read_excel('Totalidade PPLs - julho.xlsx')

#PRIMEIRA ETAPA: FAZER O MERGE E CRIAR DATFRAME COM O RESULTADO
#PASSO 01: fazer o merge dos dois dataframes pelas colunas NOME  e UNIDADE
dfCadastrosValidos = pd.merge(dfCadastrosBase, dfCadastrosSeres, on=['NOME', 'UNIDADE'])

#PASSO 02: retirar os cadastros duplicados pelo Nome e Nome da Mae
dfCadastrosValidos = dfCadastrosValidos.drop_duplicates(subset=['NOME', 'MAE'])

#SEGUNDA ETAPA: CRIAR O DATAFRAME DE CADASTROS DUPLICADOS
dfCadastrosDuplicados = dfCadastrosBase[dfCadastrosBase.duplicated(subset=['NOME', 'DATA DE NASCIMENTO'], keep=False)]

#TERCEIRA ETAPA: CRIAR DATA FRAME DE CADASTROS A ANALISAR
#PASSO 01: Realizando novamente o primeiro merge com indicador true
df_merge = pd.merge(dfCadastrosBase, dfCadastrosSeres, on=['NOME', 'UNIDADE'], how='outer', indicator=True)

# Filtrando as linhas que não foram encontradas em ambos os dataframes
dfCadastrosParaAnalisar = df_merge[df_merge['_merge'] != 'both']

#QUARTA ETAPA: CRIAR ARQUIVO EXCEL COM AS ABAS DOS RESULTADOS
with pd.ExcelWriter('DGASP - Cadastros julho.xlsx', engine='openpyxl', mode='w') as writer:
    dfCadastrosBase.to_excel(writer, sheet_name='Cadastros e-SUS', index=False)
    dfCadastrosSeres.to_excel(writer, sheet_name='Cadastros SEAP', index=False)
    dfCadastrosValidos.to_excel(writer, sheet_name='Cadastros Válidos', index=False)
    dfCadastrosDuplicados.to_excel(writer, sheet_name='Cadastros Duplicados', index=False)
    dfCadastrosParaAnalisar.to_excel(writer, sheet_name='Cadastros Para Analisar', index=False)
```
