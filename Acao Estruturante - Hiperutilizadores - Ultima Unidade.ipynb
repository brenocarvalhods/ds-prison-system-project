```python
import pandas as pd

#Criar Data Frame do arquivo de atendimentos
dfAtendimentosRealizados = pd.read_excel('Atendimentos 2023 - Hiper.xlsx')

#Fazendo a limpeza dos dados
dfAtendimentosRealizados = dfAtendimentosRealizados.drop_duplicates(subset=['NOME', 'CNS', 'CPF', 'DT. ATENDIMENTO', 'CBO'])

#Convertendo tudo para maiusculo
dfAtendimentosRealizados = dfAtendimentosRealizados.applymap(lambda x: x.upper() if isinstance(x, str) else x)
```

```python
# Converter a coluna de data para o formato datetime, se ainda não estiver
dfAtendimentosRealizados['DT. ATENDIMENTO'] = pd.to_datetime(dfAtendimentosRealizados['DT. ATENDIMENTO'])

# Encontrar a última data de atendimento para cada pessoa
df_ultimo_atendimento = dfAtendimentosRealizados.sort_values(by='DT. ATENDIMENTO').groupby('NOME').tail(1)

# Filtrar o DataFrame original para manter apenas os atendimentos na última unidade para cada pessoa
df_filtrado = dfAtendimentosRealizados.merge(df_ultimo_atendimento[['NOME', 'UNIDADE']], on=['NOME', 'UNIDADE'])

# Contar o número de atendimentos para cada pessoa na última unidade
df_contagem = df_filtrado.groupby(['NOME', 'UNIDADE', 'CBO']).size().reset_index(name='Quantidade de Atendimentos')

print(df_contagem)
```

```python
# Filtrar os atendimentos para os CBOs específicos
cbo_especificos = ['MÉDICO']

# Filtrar o DataFrame para manter apenas as linhas onde a coluna 'CBO' contém a palavra 'MÉDICO'
df_cbo_filtrado = df_contagem[df_contagem['CBO'].str.contains('MÉDICO', case=False, na=False)]

# Filtrar as pessoas com mais de 8 atendimentos
dfHiperutilizadoresMedicos = df_cbo_filtrado[df_cbo_filtrado['Quantidade de Atendimentos'] > 8]


print(dfHiperutilizadoresMedicos)
```

```python
# Filtrar os atendimentos para os CBOs específicos
cbo_especificos = ['ENFERMEIRO']

# Filtrar o DataFrame para manter apenas as linhas onde a coluna 'CBO' contém a palavra 'ENFERMEIRO'
df_cbo_filtrado = df_contagem[df_contagem['CBO'].str.contains('ENFERMEIRO', case=False, na=False)]

# Filtrar as pessoas com mais de 8 atendimentos
dfHiperutilizadoresEnfermeiros = df_cbo_filtrado[df_cbo_filtrado['Quantidade de Atendimentos'] > 8]

print(dfHiperutilizadoresEnfermeiros)
```

```python
# Filtrar os atendimentos para os CBOs específicos
cbo_especificos = ['PSICÓLOGO']

# Filtrar o DataFrame para manter apenas as linhas onde a coluna 'CBO' contém a palavra 'PSICÓLOGO'
df_cbo_filtrado = df_contagem[df_contagem['CBO'].str.contains('PSICÓLOGO', case=False, na=False)]

# Filtrar as pessoas com mais de 8 atendimentos
dfHiperutilizadoresPsicologos = df_cbo_filtrado[df_cbo_filtrado['Quantidade de Atendimentos'] > 8]

print(dfHiperutilizadoresPsicologos)
```

```python
# Filtrar os atendimentos para os CBOs específicos
cbo_especificos = ['ASSISTENTE SOCIAL']

# Filtrar o DataFrame para manter apenas as linhas onde a coluna 'CBO' contém a palavra 'ASSISTENTE SOCIAL'
df_cbo_filtrado = df_contagem[df_contagem['CBO'].str.contains('ASSISTENTE SOCIAL', case=False, na=False)]

# Filtrar as pessoas com mais de 8 atendimentos
dfHiperutilizadoresAssistenteSocial = df_cbo_filtrado[df_cbo_filtrado['Quantidade de Atendimentos'] > 8]

print(dfHiperutilizadoresAssistenteSocial)
```

```python
# Criando arquivo e abas com os resultados dos Data Frames
with pd.ExcelWriter('DGASP - Hiperutilizadores 2023 - Ultima Unid.xlsx', engine='openpyxl', mode='w') as writer:
    dfHiperutilizadoresMedicos.to_excel(writer, sheet_name='Atend. Por Médicos', index=False)
    dfHiperutilizadoresEnfermeiros.to_excel(writer, sheet_name='Atend. Por Enfermeiros', index=False)
    dfHiperutilizadoresPsicologos.to_excel(writer, sheet_name='Atend. Por Psicólogos', index=False)
    dfHiperutilizadoresAssistenteSocial.to_excel(writer, sheet_name='Atend. Por Assist. Sociais', index=False)
```
